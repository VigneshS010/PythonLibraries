# Basic Graph

```python
# Create an 1D array

x = [1,2,3,4]
y = [4,5,6,7]

# adding- color
# line width -can be changable
# marker - are used to mark the points
# markder edge color
# linestyle - changing the line type

plt.plot(x,y ,label='Line 1',color = 'green' ,linestyle = '--',linewidth=2 , marker='*', markersize = 10 , markeredgecolor = 'red')

# SHORT HABD NOTATION
# fmt = '[color][marker][line]'
# plt.plot(x, y, 'b^--', label = 'Line 1')

# We can add Font name and Font size

plt.title("My First Graph", fontdict={'fontname':'fantasy', 'fontsize':15})
plt.xlabel("X Axis", fontdict = {'fontname': 'fantasy'})
plt.ylabel("Y Axis")

# Put our own x and y axis points
plt.xticks([1,2,3,4,5])
plt.yticks([4,5,6,7,8])

# adding legend
plt.legend()

plt.show()
```

# Load Fifa Data - Histogram

```python
import pandas as pd

fifa = pd.read_csv('fifa_data.csv')
```

```python
fifa.head(3)
```

## **Histogram**

```python
bins = [50,60,70,80,90,100]
plt.hist(fifa.Overall, bins=bins , color = '#abcdef')

plt.xlabel('Skill Level')
plt.ylabel('Number of Players')
plt.title('Distribution of Player skill in FIFA')

plt.xticks(bins)
plt.show()

plt.savefig('histogram.png',dpi = 300)
```

## **Pie** **Chart**

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left','Right']
colors = ['#abcdef','#fedeed']
plt.pie([left,right], labels = labels, colors = colors, autopct = '%.2f %%' )
plt.title('Foot Preference of FIFA Players', fontdict={'fontsize':15,'fontname':'fantasy'})
plt.show()

plt.savefig('pie.png',dpi = 300)
```

```python
fifa.Weight
```

```python
# Get the weight data seperately
fifa.Weight = [int(i.strip('lbs')) if type(i)==str else i for i in fifa.Weight]
#If the inout are not only contains str value


plt.style.use('ggplot')

# Defining 5 diff categories
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight >= 200].count()[0]

weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ['Under 125','125-150','150-175','175-200','Over 200']

# we can set percentage distance
# Explode cut the graph
explode = (.4,.2,0,0,.4)
plt.pie(weights, labels = labels , autopct = '% .2f %%', pctdistance = 0.8, explode=explode)

plt.title('Weight of FIFA Players (in lbs)', fontdict={'fontsize':15,'fontname':'fantasy'})
plt.show()

plt.savefig('pie2.png',dpi = 300)
```

# Box Plot

```python
# before we chamged the style , now we need to set the default style
# ggplot style look weird here
plt.style.use('default')
plt.figure(figsize = (5,8))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['FC Barcelona','Real Madrid','New England revolution']

boxes = plt.boxplot([barcelona, madrid,revs], labels= labels, patch_artist = True)
for box in boxes['boxes']:
  # Set edge color
  box.set(color = '#abcdef', linewidth = 2)

  # Set fill color
  # If we use face color we need to set patch_artist = True
  box.set(facecolor = '#123456')


plt.title('Professional Soccer Team', fontdict={'fontsize':15,'fontname':'fantasy'})
plt.ylabel('Skill Level')

plt.savefig('box.png',dpi = 300)

plt.show()
```
