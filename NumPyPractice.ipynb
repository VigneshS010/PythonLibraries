# Linear Algebra

```python
a = np.ones((2,3))
print(a)

b = np.full((3,2),3)
print(b)

# These are in diff dimensions so direct mul wont work
# So we use matrix mul
print("Matrix Multiplication Value")
np.matmul(a,b)
```

```python
# Finding the determinant

c = np.identity(3) # identity 3 X 3 matrix
np.linalg.det(c)
```

## Reorganizing Arrays

```python
before = np.array([[1,2,3,4],[5,6,7,8]])
print(before)

after = before.reshape((8,1))
print(after)
```

```python
after = before.reshape((2,2,2))
print(after)
```

```python
#Vertically stacking vectors

v1 = np.array([1,2,3,4])
v2 = np.array([5,6,7,8])

np.vstack([v1,v2,v1,v2])
```

```python
#Horizontal stacking vector

h1 = np.ones((2,4))
h2 = np.ones((2,2))

np.hstack((h1,h2))
```
